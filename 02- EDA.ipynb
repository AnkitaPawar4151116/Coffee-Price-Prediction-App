```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds= pd.read_csv('cleand_coffee_data.csv')
ds
```

# 1. Distribution of Coffee Price

```python
plt.figure(figsize=(8, 5))
sns.histplot(ds['Price'], kde=True, color='green')
plt.title("Distribution of Coffee Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Interpretation:
This plot helps you understand how coffee prices are distributed. Look for skewness or multiple peaks. For example, a right-skew may suggest premium coffees are less common.



# 2. Count of Different Coffee Sizes

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=ds, x='Size', palette='muted')
plt.title("Count of Coffee Sizes")
plt.xlabel("Size")
plt.ylabel("Count")
plt.show()
```

Interpretation:
This chart shows how many small, medium, and large coffees are sold. If 'Medium' dominates, it could suggest itâ€™s the most preferred option.

# 3. Boxplot of Price by Coffee Type

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=ds, x='Coffee_Type', y='Price', palette='pastel')
plt.title("Price Variation by Coffee Type")
plt.xlabel("Coffee Type")
plt.ylabel("Price")
plt.show()
```

Interpretation:
This shows price range and outliers for Arabica vs Robusta. If one has a wider range or higher median, it may be a premium type.

# 4. Coffee Temperature by Milk Type

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=ds, x='Milk_Type', y='Temperature', palette='Set2')
plt.title("Temperature vs Milk Type")
plt.xticks(rotation=45)
plt.xlabel("Milk Type")
plt.ylabel("Temperature (Â°C)")
plt.show()
```

Interpretation:
Check if plant-based milks (like soy or almond) are served hotter or cooler compared to whole milk. This can affect taste perception.

#  5. Relationship Between Shots and Price

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=ds, x='Shots', y='Price', hue='Size', palette='coolwarm')
plt.title("Price vs Espresso Shots")
plt.xlabel("Number of Shots")
plt.ylabel("Price")
plt.legend(title="Size")
plt.show()
```

Interpretation:
Shows whether more espresso shots lead to higher prices and if cup size plays a role. Helpful for pricing strategies.

# 6. Most Popular Flavor

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=ds, x='Flavor', order=ds['Flavor'].value_counts().index, palette='Spectral')
plt.title("Most Preferred Coffee Flavors")
plt.xlabel("Flavor")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Interpretation:
This graph shows the popularity of different coffee flavors (e.g., Vanilla, Hazelnut, Caramel). The taller the bar, the more frequently that flavor was ordered.

# 7. Cup Quality vs Price

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=ds, x='Cup_Quality', y='Price', palette='cool')
plt.title("Cup Quality vs Coffee Price")
plt.xlabel("Cup Quality")
plt.ylabel("Price")
plt.show()
```

Interpretation:
This plot helps you understand whether better cup quality (like taste or presentation) leads to higher prices. Look for differences in medians and outliers across quality levels.
